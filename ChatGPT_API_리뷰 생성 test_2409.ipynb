```python
import openai

def generate_fake_review():
    prompt = (
        "Write a short, positive product review in Korean using 5-10 tokens. "
        "The review should be about an air purifier and include emotional expressions, "
        "talk about the price, and overall satisfaction. "
        "Don't use emoticons."
    )
    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=100,
        temperature=0.7
    )
    return response.choices[0].message['content'].strip()

# 가짜 리뷰 생성 및 출력
review = generate_fake_review()
print("Generated review:", review)
```

## 3-5. 문장이 잘리는 것 방지하고자 토큰(길이) 수 늘림 100토큰 이상(최대 400 토큰)

```python
import openai
import random

# 모델의 토큰당 가격을 정의
price_per_1m_input_tokens = 0.30
price_per_1m_output_tokens = 1.20

def generate_fake_review():
    prompts = [
        "Write a product review in Korean with more than 100 tokens. The review should be about an air purifier but try not to use the word 'air purifier'. It should be mostly positive but can include some negative nuances. Do not use emoticons. Include various aspects like price, quality, delivery, and packaging.",

        "Compose a Korean review of an air purifier using more than 100 tokens. Try to avoid using the word 'air purifier'. The review should be generally positive with some slight negative tones. Emoticons are not allowed. Discuss aspects such as price, quality, shipping, and packaging.",

        "Create a short review in Korean with more than 100 tokens about an air purifier. Avoid using the word 'air purifier'. The tone should be mostly positive but may contain a few negative nuances. Only use text, no emoticons. Mention different aspects like price, quality, delivery, and packaging.",

        "Generate a Korean product review in more than 100 tokens. The review should focus on an air purifier but avoid using the word 'air purifier'. It should be largely positive, though some negative nuances are acceptable. Emoticons should be excluded. Include comments on price, quality, shipping, and packaging."
    ]
    prompt = random.choice(prompts)
    max_tokens = 400  # Ensure a longer token limit to complete sentences

    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=max_tokens,
        temperature=0.8
    )
    # 사용된 입력 및 출력 토큰 수 반환
    return response.choices[0].message['content'].strip(), response.usage.prompt_tokens, response.usage.completion_tokens

# 100개의 가짜 리뷰 생성 및 저장
reviews = []
total_prompt_tokens = 0
total_completion_tokens = 0
for _ in range(100):
    review, prompt_tokens, completion_tokens = generate_fake_review()
    reviews.append(review)
    total_prompt_tokens += prompt_tokens
    total_completion_tokens += completion_tokens

# 총 요금 계산
total_input_cost = (total_prompt_tokens / 1_000_000) * price_per_1m_input_tokens
total_output_cost = (total_completion_tokens / 1_000_000) * price_per_1m_output_tokens
total_cost = total_input_cost + total_output_cost

# 생성된 리뷰 출력 및 총 요금 출력
for idx, review in enumerate(reviews):
    print(f"Review {idx + 1}: {review}")

print(f"\nTotal input tokens used: {total_prompt_tokens}")
print(f"Total output tokens used: {total_completion_tokens}")
print(f"Estimated total cost: ${total_cost:.4f}")
```

# 4. 리뷰 특성 파악 후 생성

```python
import pandas as pd
```

```python
import pandas as pd
import openai

# Excel 파일 읽기
file_path = 'danawa_crawling_airpurifier_reviews_result_240610.xlsx'  # 여기에 파일 경로를 입력하세요
df = pd.read_excel(file_path)

# H열의 리뷰 데이터 가져오기 (H2 셀부터 시작)
reviews = df.iloc[1:, 7].tolist()  # H열에서 두 번째 셀부터 가져옴

def analyze_review_characteristics(reviews, max_tokens=15000):
    combined_reviews = ""
    for review in reviews:
        if len(combined_reviews) + len(review) + 2 > max_tokens:  # 2는 줄바꿈과 구분자
            break
        combined_reviews += review + "\n"

    prompt = f"다음 리뷰의 특성을 분석해주세요:\n{combined_reviews}\n특성을 요약해 주세요."

    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",  # 모델 이름을 업데이트
        messages=[{"role": "user", "content": prompt}]
    )

    return response['choices'][0]['message']['content']

```

```python
# 리뷰 특성 분석
review_characteristics = analyze_review_characteristics(reviews)
print("리뷰 특성 요약:", review_characteristics)

```

```python
# 추가적으로 구조와 공통 단어 분석
def analyze_reviews_details(reviews):
    characteristics = {
        "total_reviews": len(reviews),
        "average_length": 0,
        "structures": {"short": 0, "medium": 0, "long": 0},
        "common_words": {}
    }

    all_words = []

    for review in reviews:
        length = len(review)
        characteristics["average_length"] += length
        all_words.extend(review.split())

        # 구조 분석
        if length <= 100:
            characteristics["structures"]["short"] += 1
        elif length <= 200:
            characteristics["structures"]["medium"] += 1
        else:
            characteristics["structures"]["long"] += 1

    # 평균 길이 계산
    characteristics["average_length"] /= characteristics["total_reviews"]

    # 가장 흔한 단어 분석
    for word in all_words:
        word = word.lower()  # 대소문자 구분하지 않기 위해
        if word in characteristics["common_words"]:
            characteristics["common_words"][word] += 1
        else:
            characteristics["common_words"][word] = 1

    # 가장 많이 등장한 단어 10개 선택
    characteristics["common_words"] = dict(sorted(characteristics["common_words"].items(), key=lambda x: x[1], reverse=True)[:10])

    return characteristics
```

```python

# 리뷰의 추가 특성 분석
detailed_characteristics = analyze_reviews_details(reviews)
print("리뷰 특성 상세:", detailed_characteristics)
```

### 가짜리뷰 생성

```python
# 가짜 리뷰 생성 함수
def generate_fake_reviews(review_characteristics, detailed_characteristics, num_reviews=1000):
    fake_reviews = []

    # 원하는 개수만큼 반복하여 리뷰 생성
    for _ in range(num_reviews // 10):  # 10개씩 요청
        prompt = (
            f"다음 특성을 고려하여 가짜 리뷰를 다양하게 10개 생성해 주세요:\n"
            f"총 리뷰 수: {detailed_characteristics['total_reviews']}\n"
            f"평균 길이: {detailed_characteristics['average_length']:.2f}자\n"
            f"구조: {detailed_characteristics['structures']}\n"
            f"자주 등장하는 단어: {list(detailed_characteristics['common_words'].keys())}\n"
            f"리뷰 요약: {review_characteristics}\n"
            f"가짜 리뷰:"
        )

        response = openai.ChatCompletion.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}]
        )

        # 생성된 리뷰를 추가
        fake_reviews += response['choices'][0]['message']['content'].strip().split('\n')

    return fake_reviews

# 가짜 리뷰 생성
num_reviews_to_generate = 1000  # 원하는 개수
fake_reviews = generate_fake_reviews(review_characteristics, detailed_characteristics, num_reviews=num_reviews_to_generate)

# 생성된 가짜 리뷰 출력 (문장만)
for review in fake_reviews:
    print(review.strip())
```

```python
# 가짜 리뷰를 DataFrame으로 변환
fake_reviews_df = pd.DataFrame(fake_reviews, columns=["가짜 리뷰"])

# Excel 파일로 저장
output_file_path = 'generated_fake_reviews.xlsx'  # 저장할 파일 경로
fake_reviews_df.to_excel(output_file_path, index=False)
```
