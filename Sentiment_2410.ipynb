# **mecab.morphs : 1차원 리스트로 변환**
- 한 문장을 가져와 형태소로 토큰화 한 단어가 리스트로 저장됨
- 때문에 결과가 2차원 리스트(예: [["나는", "정말"], ["영화가", "정말", "재밌다"], ....]로 저장됨
- 다음 작업인 사전과 단어를 매칭 시키기 위해서는 2차원 리스트를 1차원 리스트로 변환해 줘야 함

```python
tokens = [mecab.morphs(word) for word in reviews2401]

tokens = list(map(lambda x : " ".join(x), tokens))

tokens[:10]
```

# **토큰화된 문장과 감성사전 비교**
df2401


감성사전을 단어와 극성을 변수로 한 데이터프레임 형태로 전환
- 리뷰 한 문장씩 넣어 문장 내 단어와 감성사전 단어를 매치
- 존재하면 극성값을 가져오고, 없으면 패스

- 리뷰 문장별 극성값은 문장의 포함된 단어의 극성값 합계로 계산
- (예) "나는 이 영화가 재밌고 감동적이다"라면, '재밌고'가 +1, 감동적이다 +2점으로 이 문장은 긍정적인 +3점 리뷰로 판단

```python
#감성사전을 단어와 극성을 변수로 한 데이터프레임 형태로 전환
import pandas as pd

df2401 = pd.DataFrame(columns=("review", "sentiment"))  # 리뷰별 극성을 저장하기 위한 데이터프레임 생성
idx = 0                                             # 다음 리뷰로 넘기기 위한 초기값

for token in tokens:                                # 전체 리뷰에서 문장 하나씩 가져옴
  sentiment = 0                                     # 초기 감성값 0으로 설정
  for i in range(0, len(sentiword_dic)):            # 감성사전의 모든 단어를 하나씩 선택
    if sentiword_dic.word[i] in token:              # 리뷰 문장에 감성 단어가 있는지 확인
      sentiment += int(sentiword_dic.polarity[i])   # 감성단어가 있다면 극성값 합계를 구함.
  df2401.loc[idx] = [token, sentiment]                  # 리뷰별 극성값을 데이터프레임으로 쌓음
  idx += 1                                          # 다름 리뷰 문장으로 넘어감
```
