```python
from konlpy.tag import Okt

# Okt 형태소 분석기 객체 생성
okt = Okt()
```

```python
# 한국어 텍스트 전처리 및 품사 태깅 함수
def preprocess_text(text):
    text = re.sub(r'\W+', ' ', text)  # 특수문자 제거
    okt = Okt()
    tokens = okt.pos(text, stem=True)  # 품사 태깅 및 어간 추출
    # 명사와 동사만 선택
    tokens = [word for word, pos in tokens if pos in ['Noun', 'Verb']]
    return ' '.join(tokens)
```

```python
# 모든 리뷰에 대해 전처리 수행
post_reviews = processed_reviews.apply(preprocess_text)
```

# Stopwords

```python
stopwords = ['데/NNB', '좀/MAG', '수/NNB', '등/NNB']

def remove_stopword_text(text):
    corpus = []
    for sent in text:
        modi_sent = []
        for word in sent.split(' '):
            if word not in stopwords:
                modi_sent.append(word)
        corpus.append(' '.join(modi_sent))
    return corpus
```

```python
# 모든 리뷰에 대해 전처리 수행
reviews_fin = post_reviews.apply(remove_stopword_text)
```

```python
#형태소 분리 후 post_reviews에 대해서 벡터화 (숫자화)

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(post_reviews)
```

# K-means 분류

```python
from sklearn.cluster import KMeans

# K-Means 클러스터링
num_clusters = 2  # 클러스터 수는 데이터에 따라 조정
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)
```

```python
# 각 리뷰에 할당된 클러스터 라벨
labels = kmeans.labels_

# 결과를 데이터프레임에 추가
df['Cluster'] = labels
```

```python
df_kmeans = df[['리뷰', 'Cluster']]
df_kmeans
```

```python
cluster_counts = df['Cluster'].value_counts()
print(cluster_counts)

# 0 또는 1로 구분됨
```

# IsolationForest

```python
from sklearn.ensemble import IsolationForest
# Isolation Forest를 사용한 이상치 탐지
iso_forest = IsolationForest(contamination=0.1, random_state = 100)  # 데이터의 10%를 이상치로 가정
iso_forest.fit(X)

# 각 리뷰에 대한 이상치 여부 예측
labels = iso_forest.predict(X)

# 이상치 탐지 결과를 데이터프레임에 추가
df['Outlier'] = labels

# 결과 출력 (-1은 이상치, 1은 정상 데이터)
print(df[['리뷰', 'Outlier']])
```
