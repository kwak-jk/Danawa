## 3-2. gpt-4o-mini 이용하여 리뷰 100개 출력

```python
import openai

# 모델의 토큰당 가격을 정의
price_per_1m_input_tokens = 0.30
price_per_1m_output_tokens = 1.20

def generate_fake_review():
    prompt = (
        "Write a short, positive product review in Korean using 5-10 tokens. "
        "The review should be about an air purifier and include emotional expressions, "
        "talk about the price, and overall satisfaction. "
        "Don't use emoticons."
    )
    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=100,
        temperature=0.7
    )
    # 사용된 입력 및 출력 토큰 수 반환
    return response.choices[0].message['content'].strip(), response.usage.prompt_tokens, response.usage.completion_tokens

# 100개의 가짜 리뷰 생성 및 저장
reviews = []
total_prompt_tokens = 0
total_completion_tokens = 0
for _ in range(100):
    review, prompt_tokens, completion_tokens = generate_fake_review()
    reviews.append(review)
    total_prompt_tokens += prompt_tokens
    total_completion_tokens += completion_tokens

# 총 요금 계산
total_input_cost = (total_prompt_tokens / 1_000_000) * price_per_1m_input_tokens
total_output_cost = (total_completion_tokens / 1_000_000) * price_per_1m_output_tokens
total_cost = total_input_cost + total_output_cost

# 생성된 리뷰 출력 및 총 요금 출력
for idx, review in enumerate(reviews):
    print(f"Review {idx + 1}: {review}")

print(f"\nTotal input tokens used: {total_prompt_tokens}")
print(f"Total output tokens used: {total_completion_tokens}")
print(f"Estimated total cost: ${total_cost:.4f}")
```
